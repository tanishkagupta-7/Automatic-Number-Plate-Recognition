## 🧠 Introduction
Automatic Number Plate Recognition (ANPR) is a computer vision application that identifies vehicle registration plates from images. It has use cases in surveillance, traffic monitoring, toll collection, and parking systems.

In this notebook, we:
- Load a vehicle image
- Manually crop the number plate
- Apply OCR (Optical Character Recognition)
- Extract text from the plate


```python
# Import Libraries
import cv2
import matplotlib.pyplot as plt

import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'TRUE'
import os
print(os.getcwd())

import easyocr
```

```python
## Step 1: Load the image
img = cv2.imread("no_plate.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 6))
plt.imshow(img_rgb)
plt.title("Original Image - Locate Number Plate")
plt.axis("on")
plt.grid(True)
plt.show()
```

```python
## Step 2: Manual Cropping
x1, y1 = 430, 430
x2, y2 = 780, 580
cropped = img[y1:y2, x1:x2]

cv2.imwrite("manual_crop.jpg", cropped)

plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
plt.title("✅ Final Cropped Number Plate")
plt.axis("off")
plt.show()
```

```python
## Step 3: OCR on Cropped Image
reader = easyocr.Reader(['en'])

result = reader.readtext(cropped)

if len(result) == 0:
    print("😕 No text detected.")
else:
    print("✅ Detected Text:", result[0][1])
```
